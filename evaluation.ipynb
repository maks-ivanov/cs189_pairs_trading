```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# loading model evaluation history
lstm2_raw_residual = pd.read_csv('lstm2_raw_residual_history.csv')
bilstm2_raw_residual = pd.read_csv('bilstm2_raw_residual_history.csv')
bilstm3_raw_residual =  pd.read_csv('bilstm3_raw_residual_history.csv')

lstm2_zone_feature = pd.read_csv('lstm2_zone_feature_history.csv')
bilstm2_zone_feature = pd.read_csv('bilstm2_zone_feature_history.csv')
bilstm3_zone_feature = pd.read_csv('bilstm3_zone_feature_history.csv')

lstm2_residual_features = pd.read_csv('lstm2_residual_features_history.csv')
bilstm2_residual_features = pd.read_csv('bilstm2_residual_features_history.csv')
bilstm3_residual_features = pd.read_csv('bilstm3_residual_features_history.csv')

lstm2_ts_zone_features = pd.read_csv('lstm2_ts_zone_features_history.csv')
bilstm2_ts_zone_features = pd.read_csv('bilstm2_ts_zone_features_history.csv')
bilstm3_ts_zone_features = pd.read_csv('bilstm3_ts_zone_features_history.csv')

lstm2_ts_residual_features = pd.read_csv('lstm2_ts_residual_features_history.csv')
bilstm2_ts_residual_features = pd.read_csv('bilstm2_ts_residual_features_history.csv')
bilstm3_ts_residual_features = pd.read_csv('bilstm3_ts_residual_features_history.csv')

raw_res = {'lstm2' : lstm2_raw_residual
          ,'bilstm2' : bilstm2_raw_residual
          ,'bilstm3' : bilstm3_raw_residual}

zone_feat = {'lstm2' : lstm2_zone_feature
          ,'bilstm2' : bilstm2_zone_feature
          ,'bilstm3' : bilstm3_zone_feature}

res_feat = {'lstm2' : lstm2_residual_features
           ,'bilstm2' : bilstm2_residual_features
           ,'bilstm3' : bilstm3_residual_features}

ts_res_feat = {'lstm2' : lstm2_ts_residual_features
              ,'bilstm2' : bilstm2_ts_residual_features
              ,'bilstm3' : bilstm3_ts_residual_features}
ts_zone_feat = {'lstm2' : lstm2_ts_zone_features
              ,'bilstm2' : bilstm2_ts_zone_features
              ,'bilstm3' : bilstm3_ts_zone_features}

colors = {'lstm2':'b'
         ,'bilstm2':'r'
         ,'bilstm3':'g'}

figsize = (12, 8)

datasets = {'Time series features from residual': ts_res_feat
           ,'Time series features from zone' : ts_res_feat
           ,'Raw residual' : raw_res
           ,'Residual features' : res_feat
           , 'Zone feature' : zone_feat}

all_evaluations = []

for ds in datasets.values():
    all_evaluations += ds.values()
```

```python
plt.figure(figsize=figsize)
for model_name, df in raw_res.items():
    df['categorical_accuracy'].plot(label = model_name+' train', color=colors[model_name], alpha=.7, style=':')
    df['val_categorical_accuracy'].plot(label = model_name + ' test', color=colors[model_name],alpha=.7)
plt.title('raw residual accruacy')
plt.legend()
plt.show()

# plt.figure(figsize=figsize)
# for model_name, df in raw_res.items():
#     #df['categorical_accuracy'].plot(label = model_name+' train', color=colors[model_name], alpha=.5)
#     df['val_categorical_accuracy'].plot(label = model_name + ' test', color=colors[model_name], alpha=.7)
# plt.title('raw residual test accruacy')
# plt.legend()
# plt.show()
```

```python
plt.figure(figsize=figsize)
for model_name, df in zone_feat.items():
    df['categorical_accuracy'].plot(label = model_name+' train', color=colors[model_name],style=':', alpha=.7)
    df['val_categorical_accuracy'].plot(label = model_name + ' test', color=colors[model_name], alpha=.7)
plt.title('zone feature accruacy')
plt.legend()
plt.show()

# plt.figure(figsize=figsize)
# for model_name, df in zone_feat.items():
#     #df['categorical_accuracy'].plot(label = model_name+' train', color=colors[model_name], alpha=.5)
#     df['val_categorical_accuracy'].plot(label = model_name + ' test', color=colors[model_name], alpha=.7)
# plt.title('zone feature test accruacy')
# plt.legend()
# plt.show()
```
